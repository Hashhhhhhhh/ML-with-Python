Data standardization is a preprocessing step often used in machine learning and data analysis. It involves scaling numerical features so that they have a mean of 0 and a standard deviation of 1. In simply words, it is standardizing the data to a common format or common range.

StandardScalar is the function to standardize the entire dataset.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.datasets import load_breast_cancer
import sklearn
```
