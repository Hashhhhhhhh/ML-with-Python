An imbalanced data :

The one with unequal class distribution

```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('/content/credit_data.csv')
df.head(6)
```

Typical classification problem.

```python
df.shape
```

```python
#Distribution of the two classes
df['Class'].value_counts()
```

0 : Legit transactions


1 : Fradulent transactions

```python
#Seperating legit and fraud
legit=df[df.Class==0]
fraud=df[df.Class==1]
print(legit.shape)
print(fraud.shape)
```

Under sampling technique to handle imbalanced data.

Build a small dataset containing similar distribution of legit and fraudulent transactions.

```python
legit_sample=legit.sample(n=104)
print(legit_sample.shape)
```

```python
#concatenate the two dataframes
new_df=pd.concat([legit_sample,fraud],axis=0)
new_df.shape
```
