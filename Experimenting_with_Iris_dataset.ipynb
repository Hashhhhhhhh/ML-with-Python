```python
import sklearn
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
#Visualizing through scatter plot
plt.scatter(iris.data[:,0],iris.data[:,1],c=iris.target)
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.show()
print()
plt.scatter(iris.data[:,0],iris.data[:,2],c=iris.target)
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[2])
plt.show()
print()
```

```python
plt.scatter(iris.data[:,0],iris.data[:,3],c=iris.target)
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[3])
plt.show()
print()
plt.scatter(iris.data[:,1],iris.data[:,2],c=iris.target)
plt.xlabel(iris.feature_names[1])
plt.ylabel(iris.feature_names[2])
plt.show()
print()
```

```python
#4features and target
df=pd.DataFrame(iris.data,columns=iris.feature_names)
df['target']=iris.target
df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(df[['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)']],df['target'],test_size=0.2,random_state=42)
```

```python
knn=KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train,Y_train)
```

```python
#to see model's performance- accuracy by comparing the prediction with actual results
Y_pred=knn.predict(X_test)
from sklearn.metrics import accuracy_score
accuracy_score(Y_test,Y_pred)
```

While a perfect accuracy score (of 1) is ideal, in real-world scenarios it could also indicate that your model is overfitting (memorizing the training data) or that your test data is too easy or not representative of real-world cases. Always check for data leakage or class imbalance if you see an unexpectedly perfect score.

1. Class imbalance means some classes have many more samples than others, which can bias your model. (use print(df['target'].value_counts()) to check)
